```python
# Pickling Data
with open("data/area.pickle", "rb") as fr:
    data_area = pickle.load(fr)
```

```python
data_area['mtrxpy']
```

```python
area = {}

area['area_ttl'] = data_area['mtrxpy'].loc[('ttl', 'sum'), ('ttl', 'sum')]

loc_cld = [x  in ['b1', 'f1'] for x in data_area['mtrxpy'].index.get_level_values(1)]
area['area_cld'] = sum(data_area['mtrxpy'].loc[:, ('ttl', 'sum')][loc_cld])

loc_nml = [x  in ['f2', 'f3', 'f4'] for x in data_area['mtrxpy'].index.get_level_values(1)]
area['area_nml'] = sum(data_area['mtrxpy'].loc[:, ('ttl', 'sum')][loc_nml])

area['floor_ttl'] = data_area['mtrxpy'].loc[('ttlfloor', 'sum'), ('ttl', 'sum')]

loc_cld = [x in ['b1', 'f1'] for x in data_area['mtrxpy'].loc['ttlfloor'].index]
area['floor_cld'] = sum(data_area['mtrxpy'].loc['ttlfloor'].loc[:, ('ttl', 'sum')][loc_cld])

loc_nml = [x in ['f2', 'f3', 'f4'] for x in data_area['mtrxpy'].loc['ttlfloor'].index]
area['floor_nml'] = sum(data_area['mtrxpy'].loc['ttlfloor'].loc[:, ('ttl', 'sum')][loc_nml])

area['rent_ttl'] = data_area['mtrxpy'].loc[('ttl', 'sum'), ('rent', 'sum')]

loc_cld = [x  in ['b1', 'f1'] for x in data_area['mtrxpy'].index.get_level_values(1)]
area['rent_cld'] = sum(data_area['mtrxpy'].loc[:, ('rent', 'sum')][loc_cld])

loc_nml = [x  in ['f2', 'f3', 'f4'] for x in data_area['mtrxpy'].index.get_level_values(1)]
area['rent_nml'] = sum(data_area['mtrxpy'].loc[:, ('rent', 'sum')][loc_nml])


print(f"Total area  : {area['area_ttl']:>10,.1f}")
print(f"Cold area   : {area['area_cld']:>10,.1f}")
print(f"Nomal area  : {area['area_nml']:>10,.1f}")
print(f"Total floor : {area['floor_ttl']:>10,.1f}")
print(f"Cold floor  : {area['floor_cld']:>10,.1f}")
print(f"Nomal floor : {area['floor_nml']:>10,.1f}")
print(f"Total rent  : {area['rent_ttl']:>10,.1f}")
print(f"Cold rent   : {area['rent_cld']:>10,.1f}")
print(f"Nomal rent  : {area['rent_nml']:>10,.1f}")
```

# Initial Setting

```python
astn = EmptyClass()
```

# Input Index Data

```python
tmp_col = ['idxname', 'start', 'periods']
tmp_ary = [['prjt',  '2021-12', 35],
           ['cstrn', '2022-01', 23],
           ['loan',  '2022-01', 29],
           ['sales', '2022-01', 29]]
astn.index = DataFrame(tmp_ary, columns=tmp_col)

idx = cf.PrjtIndex(idxname = list(astn.index.idxname),
                   start   = list(astn.index.start),
                   periods = list(astn.index.periods + 1),
                   freq    = 'M')
```

# Input Financing Condition Data

```python
# Equity Data

tmp_col = ['title', 'amt_ntnl', 'amt_intl']
tmp_ary = [['equity',  1_300,     1_300]]
astn.equity = DataFrame(tmp_ary, columns=tmp_col)

equity = cf.Loan(idx,
                 amt_ntnl = astn.equity.amt_ntnl[0],
                 amt_intl = astn.equity.amt_intl[0])
```

```python
# Loan Data

tmp_col = ['title', 'rnk', 'amt_ntnl', 'amt_intl', 'rate_fee', 'rate_IR']
tmp_ary = [['tra',     0,     75_000,     18_000,      0.020,     0.043],
           ['trb',     1,     22_000,     22_000,      0.041,     0.070]]
astn.loan = DataFrame(tmp_ary, columns=tmp_col).sort_values(by=['rnk'])

loan = cf.Intlz_loan(idx, idx.loan,
                     title = astn.loan.title,
                     rnk = astn.loan.rnk,
                     amt_ntnl = astn.loan.amt_ntnl,
                     amt_intl = astn.loan.amt_intl,
                     rate_fee = astn.loan.rate_fee,
                     rate_IR = astn.loan.rate_IR)

loan.arng_fee = 0.02       # Fee on balance for Tr.A
loan.feeonbal_tra = 0.002  # Arangement fee rate

print("## Financing Condition ##\n")
print(f"Total Equity amount:{sum(astn.equity.amt_ntnl):>7,.0f}")
print(f"Total loan amount:  {sum(astn.loan.amt_ntnl):>7,.0f}")
print(R(astn.loan, rate=['rate_fee', 'rate_IR']))
print(f"Arangement fee rate: {loan.arng_fee * 100:,.1f}%")
print(f"Fee on bal Tr.A    : {loan.feeonbal_tra * 100:,.1f}%")
```

```python
fee_amt = sum(astn.loan.amt_ntnl * astn.loan.rate_fee)
IR_amt = sum(astn.loan.amt_ntnl * astn.loan.rate_IR * 29 / 12)
arng_amt = sum(astn.loan.amt_ntnl * loan.arng_fee)

ttl_amt = fee_amt + IR_amt + arng_amt
allin_rate = ttl_amt / sum(astn.loan.amt_ntnl) * 12 / 29

print(f"allin_rate : {allin_rate*100:>5.2f}%")
```

### 2) Land Purchase Costs

```python
lnd = {}
```

```python
title = ('lnd', 'prchs') # land purchase cost
byname = '용지매입비'
lnd['prchs'] = 14_500
tmp_ary = [[idx.prjt[0],  1_000],
           [idx.loan[0], 13_500]]
cost.inptcst(title, byname, scdddf=tmp_ary)

print(f"Total prchs  : {lnd['prchs']:>10,.1f}")
print(f"Intl prchs   : {cost['prchs'].add_scdd[idx.prjt[0]]:>10,.1f}")
print(f"Secnd prchs  : {cost['prchs'].add_scdd[idx.loan[0]]:>10,.1f} ")
```

```python
title = ('lnd', 'brkrg') # land brokerage fee
byname = '부동산중개비용'
lnd['brkrg'] = lnd['prchs'] * 0.009
intlcst = 39
tmp_ary = [[idx.prjt[0],  intlcst],
           [idx.loan[0],  lnd['brkrg'] - intlcst]]
cost.inptcst(title, byname, scdddf=tmp_ary)

print(f"Total brkrg  : {lnd['brkrg']:>10,.1f}")
print(f"Intl brkrg   : {intlcst:>10,.1f}")
print(f"Secnd brkrg  : {cost['brkrg'].add_scdd[idx.loan[0]]:>10,.1f}")
```

```python
title = ('lnd', 'nhbond') # national housing bond
byname = '국민주택채권'
매입률 = 0.045
본인부담률 = 0.055
lnd['nhbond'] = lnd['prchs'] * 매입률 * 본인부담률
tmp_ary = [[idx.loan[0], lnd['nhbond']]]
cost.inptcst(title, byname, scdddf=tmp_ary)

print(f"Total nhbond : {cost['nhbond'].add_scdd[idx.loan[0]]:10,.1f}")
```

```python
title = ('lnd', 'jdclcst') # judicial scrivener cost
byname = '법무사비용'
수수료율 = 0.001
lnd['jdclcst'] = lnd['prchs'] * 수수료율
tmp_ary = [[idx.loan[0], lnd['jdclcst']]]
cost.inptcst(title, byname, scdddf=tmp_ary)

print(f"Total jdclcst : {cost['jdclcst'].add_scdd[idx.loan[0]]:10,.1f}")
```

```python
title = ('lnd', 'tax_aqstn') # land acquisition and registration tax
byname = '취등록세'
취득세율 = 0.040
농특세율 = 0.002
교육세율 = 0.004
lnd['tax_aqstn'] = sum(cost['lnd'].amt) * (취득세율 + 농특세율 + 교육세율)
tmp_ary = [[idx.loan[0], lnd['tax_aqstn']]]
cost.inptcst(title, byname, scdddf=tmp_ary)

print(f"Total tax_aqstn : {cost['tax_aqstn'].add_scdd[idx.loan[0]]:10,.1f}")
```

```python
lnd['sum'] = sum(cost['lnd'].amt)

print(f"Total land cost : {lnd['sum']:10,.1f}")
```

```python
cost.lnd = lnd
```

### 3) Construction Costs

```python
cstrn = {}
cstrn['dmltn'] = 35 + 1_180 # 철거비: 3,550평 x 342천원
cstrn['civil'] = 8_490 # 토목공사: 16,327평 x 520천원
cstrn['nml_fclt'] = area['area_nml'] * 2.66 # 상온시설: 9,311평 x 2,660천원
cstrn['cld_fclt'] = area['area_cld'] * 3.66 # 저온시설: 8,365평 x 3,660천원
cstrn['ttl_fclt'] = cstrn['nml_fclt'] + cstrn['cld_fclt']

cstrn['sum'] = 65_000 # 철거, 토목, 건축공사 통합 650억원으로 설정

retention_rate = 0.1
cstrn['intl'] = 35
cstrn['sum_prd'] = cstrn['sum'] * (1-retention_rate) - cstrn['intl']
cstrn['sum_rttn'] = cstrn['sum'] * retention_rate

print(f"Total construction cost:   {cstrn['sum']:>10,.0f}")
print(f"Initial construction cost: {cstrn['intl']:>10,.0f}")
print(f"Period construction cost:  {cstrn['sum_prd']:>10,.0f}")
print(f"Retained construction cost:{cstrn['sum_rttn']:>10,.0f}")
```

```python
title = ('cstrn', 'cstrn') # construction cost
byname = '공사비'
tmp_idx = idx.cstrn.index
tmp_len = len(idx.cstrn)
tmp_ary = [[idx.prjt[0], cstrn['intl']]]
tmp_ary.extend([[x, cstrn['sum_prd'] / tmp_len] for x in tmp_idx])
tmp_ary.append([idx.prjt[-1], cstrn['sum_rttn']])

cost.inptcst(title, byname, scdddf=tmp_ary)
```

```python
cost.cstrn = cstrn
```
